```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
Choc = pd.read_csv('C:\\Users\\cocob\\Downloads\\flavors_of_cacao.csv')
```

```python
Choc.rename(columns={'Specific Bean Origin\nor Bar Name':'Specific_Bean', 'Review\nDate':'Review Date', 'Cocoa\nPercent':'Cocoa Percent', 'Company\nLocation':'Company Location', 'Bean\nType':'Bean Type', 'Broad Bean\nOrigin':'Bean Origin'}, inplace=True)
```

```python
Choc['Cocoa Percent']= Choc['Cocoa Percent'].str.rstrip('%').astype('float') / 100.0
```

```python
Choc.columns.values
```

### Part 1

```python
Choc1= Choc.head()
Choc2= Choc1[['Rating']]
Choc2.plot(kind= 'bar', color ='aqua')
plt.show()

Choc2.plot(color ='pink')
plt.show()

Choc2.hist(color ='turquoise')
plt.show()

Choc2.boxplot()
plt.show()
```

### Part 2

```python
colors= ('Aqua', 'Pink')
Choc3= Choc.groupby(['Rating'])['Cocoa Percent'].sum().reset_index()
Choc4= Choc3.sort_values(by= 'Cocoa Percent', ascending=False)

Choc4.plot(kind= 'Bar', color= colors)
plt.show()

sns.set(style='darkgrid')
Choc5 = sns.factorplot(x='Rating', y='Cocoa Percent', data=Choc, size=10, kind='bar', palette='pastel')
Choc5.despine(left=True)
Choc5.set_ylabels('Cocoa Percent')
Choc5.set_xlabels('Rating')
plt.title('Cocoa Rating')
plt.show()

```

```python
Choc5 = sns.lmplot(y='Cocoa Percent', x='Rating', data=Choc, fit_reg=True)
Choc5.set_ylabels('Cocoa Percent')
Choc5.set_xlabels('Rating')
plt.title('Cocoa Rating')
plt.show()
```

### Part 3

```python
Choc6= Choc.head(50)
sns.set(style= 'whitegrid')
Choc7= sns.factorplot(x='Company Location', y='Rating', data=Choc6, size=8, kind='point', palette='pastel')
plt.title('Company Rating by Location Type')
plt.show()

sns.set(style='darkgrid')
Choc8= sns.factorplot(x= 'Company Location', y='Rating', data=Choc6, size=8, kind= 'bar', palette='BuPu')
Choc5.despine(left=True)
Choc5.set_ylabels('Company Location')
Choc5.set_xlabels('Rating')
plt.title('Company Rating by Location Type')
plt.show()

sns.set(style='whitegrid')
Choc9= sns.boxplot(x='Company Location', y='Rating', data=Choc6, palette='Spectral') 
plt.title('Company Rating by Locatin Type')
plt.show()

Choc6.plot(x= 'Company Location', y='Rating', color= 'red')
plt.show()

Choc10= sns.swarmplot(x='Company Location', y='Rating', data=Choc6, size=8, palette='PuBuGn')
plt.title('Company Rating by Location Type')
plt.show()

Choc11= sns.violinplot(x='Company Location', y='Rating', data=Choc6, size=8, palette='pastel')
plt.title('Company Rating by Locatio Type')
plt.show()
```
